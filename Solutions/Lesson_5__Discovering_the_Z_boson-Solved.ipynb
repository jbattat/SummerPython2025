```python
#Imports etc
import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

Last time we may have seen hints of new, heavy particles by looking at oposite sign di-leptons.  Because of momentum conservation we should be able to see the true smoking-gun, a mass peak.  If we can see a clear mass peak, we have found evidence of a new heavy particle in collider data.  This is how the Higgs boson, and numerous others, were discovered.



```python
# Exercise : load your data samples.  Calculate the invariant mass of the
#            highest energy lepton pairs in each sub-category

#            Histogram the invariant masses.  What do you see?

EE=numpy.loadtxt("ee.txt")
EMu=numpy.loadtxt("emu.txt")
MuMu=numpy.loadtxt("mumu.txt")

InvMassesEE=[]
InvMassesEMu=[]
InvMassesMuMu=[]

for ev in EE:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        InvMassesEE.append(InvMass)

for ev in EMu:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        InvMassesEMu.append(InvMass)


for ev in MuMu:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        InvMassesMuMu.append(InvMass)
```

```python
plt.figure(figsize=(5,5))
plt.hist(InvMassesMuMu,bins=numpy.arange(0,300,1),histtype='step',label="mu mu")
plt.hist(InvMassesEMu,bins=numpy.arange(0,300,1),histtype='step',label="e mu")
plt.hist(InvMassesEE,bins=numpy.arange(0,300,1),histtype='step',label="e e")
plt.xlabel("Invariant Mass",fontsize=14)
plt.ylabel("Counts",fontsize=14)
plt.legend(loc=1,fontsize=14)
```

```python
InvMassesEESameSign=[]
InvMassesEMuSameSign=[]
InvMassesMuMuSameSign=[]

InvMassesEEOppSign=[]
InvMassesEMuOppSign=[]
InvMassesMuMuOppSign=[]

for ev in EE:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        if(ev[6]*ev[12]==1):
            InvMassesEESameSign.append(InvMass)
        else:
            InvMassesEEOppSign.append(InvMass)
            
for ev in EMu:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        if(ev[6]*ev[12]==1):
            InvMassesEMuSameSign.append(InvMass)
        else:
            InvMassesEMuOppSign.append(InvMass)

for ev in MuMu:
    Total_E  = ev[5]+ev[11]
    Total_px = ev[2]+ev[8]
    Total_py = ev[3]+ev[9]
    Total_pz = ev[4]+ev[10]
    
    InvMass2=Total_E**2-Total_px**2-Total_py**2-Total_pz**2
    if(InvMass2>0):
        InvMass=pow(InvMass2,0.5)
        if(ev[6]*ev[12]==1):
            InvMassesMuMuSameSign.append(InvMass)
        else:
            InvMassesMuMuOppSign.append(InvMass)
```

```python
plt.figure(figsize=(5,5))
plt.hist(InvMassesMuMuSameSign,bins=numpy.arange(0,300,1),histtype='step',label="mu mu same")
plt.hist(InvMassesMuMuOppSign,bins=numpy.arange(0,300,1),histtype='step',label="mu mu opp")
plt.xlabel("Invariant Mass",fontsize=14)
plt.ylabel("Counts",fontsize=14)
plt.legend(loc=1,fontsize=14)

plt.figure(figsize=(5,5))
plt.hist(InvMassesEESameSign,bins=numpy.arange(0,300,5),histtype='step',label="e e same")
plt.hist(InvMassesEEOppSign,bins=numpy.arange(0,300,5),histtype='step',label="e e opp")
plt.xlabel("Invariant Mass",fontsize=14)
plt.ylabel("Counts",fontsize=14)
plt.legend(loc=1,fontsize=14)

plt.figure(figsize=(5,5))
plt.hist(InvMassesEMuSameSign,bins=numpy.arange(0,300,5),histtype='step',label="e mu same")
plt.hist(InvMassesEMuOppSign,bins=numpy.arange(0,300,5),histtype='step',label="e mu opp")
plt.xlabel("Invariant Mass",fontsize=14)
plt.ylabel("Counts",fontsize=14)
plt.legend(loc=1,fontsize=14)
```
